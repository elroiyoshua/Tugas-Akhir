```python
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
import gensim
from collections import Counter
import ast
```

```python
df = pd.read_csv("merged_data_all_labeled.csv")


# Tweet depresif bernilai 1
df["label"] = df["label"].replace("Positif", 1)
# Tweet non depresif bernilai 0
df["label"] = df["label"].replace("Negatif", 0)
df
```

```python
def convert_string_to_list(string_representation):
    return ast.literal_eval(string_representation)

df['full_text'] = df['full_text'].apply(lambda x: convert_string_to_list(x))
df
```

## Splitting Dataset

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X = df["full_text"].tolist()
```

```python
X
```

```python
y = df["label"].tolist()
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

# Word2Vec CBOW

```python
cbow_model_path = "cbow_wiki_word2vec_gensim3.model"
```

```python
model_skipgram = gensim.models.Word2Vec.load(cbow_model_path)
```

```python
print(model_skipgram.wv['ko'])
```

```python
similarity = model_skipgram.wv.most_similar("indonesia", topn=10)
print(similarity)
```

```python
# Ubah semua token dalam data menjadi representasi vektor dengan Word2Vec
def sent_vec(sent):
    vector_size = model_skipgram.wv.vector_size
    wv_res = np.zeros(vector_size)
    # print(wv_res)
    ctr = 1
    for w in sent:
        if w in model_skipgram.wv:
            ctr += 1
            wv_res += model_skipgram.wv[w]
    wv_res = wv_res/ctr
    return wv_res
```

```python
X_trainvec =[]
X_testvec = []
```

```python
for i in range(len(X_train)):
    X_trainvec.append(sent_vec(X_train[i]))
X_trainvec
```

```python
for i in range(len(X_test)):
    X_testvec.append(sent_vec(X_test[i]))
X_testvec
```

```python
print("Jumlah data X_train :", len(X_trainvec))
print("Jumlah data X_test :", len(X_testvec))
```

```python
print("Jumlah data y_train :", len(y_train))
print("Jumlah data y_test :", len(y_test))
```

# Implementation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

from sklearn import model_selection
from sklearn import preprocessing
from sklearn.metrics import classification_report, confusion_matrix
```

## Decision Tree

```python
hasil_params=[]
hasil_akurasi=[]
hasil_all=[]
```

```python
# Standarisasi fitur yang digunakan
scaler = preprocessing.StandardScaler()
X_train_scaled = scaler.fit_transform(X_trainvec)
X_test_scaled = scaler.transform(X_testvec)

# Hyperparameter yang akan digunakan untuk menentukan parameter terbaik
param_grid = {'criterion': ['gini'],
              'max_depth': [None],
              }

# Decision Tree Classifier
dt_classifier = DecisionTreeClassifier()

# Grid Search untuk mencari parameter terbaik
grid_search = model_selection.GridSearchCV(dt_classifier, param_grid, cv=5)
grid_search.fit(X_train_scaled, y_train)

# Parameter terbaik
best_params_dt = grid_search.best_params_

# Train Decision Tree menggunakan parameter terbaik
dt_classifier = DecisionTreeClassifier(**best_params_dt)
dt_classifier.fit(X_train_scaled, y_train)

# Evaluasi model yang dibuat
X_test_scaled = scaler.transform(X_testvec)
y_pred = dt_classifier.predict(X_test_scaled)
accuracy = dt_classifier.score(X_test_scaled, y_test)

# Akurasi model
print(best_params_dt)
print("Accuracy:", accuracy)

# Calculate precision, recall, and F1-score
classification_rep = classification_report(y_test, y_pred)
print("Classification Report:\n", classification_rep)
# Extract F1-score from the classification report
f1_score_value = float(classification_rep.split()[-2])

# Print F1-score
print("F1 Score:", f1_score_value)

# Calculate confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", confusion_mat)
hasil_params.append(best_params_dt)
hasil_akurasi.append(accuracy)
hasil_all.append(f1_score_value)
```
