````python
import gensim
````

````python
file_wiki = "idwiki-latest-pages-articles.xml.bz2"
file_idwiki = "idwiki.txt"
wiki_indo = gensim.corpora.WikiCorpus(file_wiki, dictionary={})
````

# **Read Wikipedia Dump**




```
WikipediaDump.bz2 -> WikipediaText.txt
```



````python
with open(file_idwiki, "w", encoding='utf-8') as f:
  i = 0
  for text in wiki_indo.get_texts():
    text = ' '.join(text)
    text = text.lower()
    f.write(text + '\n')
    i += 1
print("Total text :", str(i))
````

# **Word2Vec Skipgram Model**

## Build Model

**WARNING : PROSES RUNNING LAMA; JALANKAN HANYA SEKALI SAJA âš **

````python
skipgram_model_path = "skipgram_wiki_word2vec_gensim3.model"
````

````python
sentences = gensim.models.word2vec.LineSentence(file_idwiki)
skipgram_model = gensim.models.word2vec.Word2Vec(sentences, vector_size=300, window=5, min_count=1, workers=4, sg=1)
skipgram_model.save(skipgram_model_path)
````

## Testing

````python
skipgram_model = gensim.models.Word2Vec.load(skipgram_model_path)
````

````python
print(skipgram_model.wv['indonesia'])
````

````python
similarity = skipgram_model.wv.most_similar("komputer", topn=10)
print(similarity)
````
